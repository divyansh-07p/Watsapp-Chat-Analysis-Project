```python
import re #regular expression module
import pandas as pd #pandas lib
```

```python
f=open("WhatsApp Chat with ARPAN Core Team 2025.txt","r",encoding="utf-8")#utf-8 is encoding format that supports various characters, including special symbols and non-English text (like Hindi or emojis).
data=f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern,data)[1:]#pehli line bhi aa rhi thi isliye use hatane ke liye ye [1:] lagaya hta ke dekh lo
print(messages)
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})
#convert message_datatype
df['message_date']=pd.to_datetime(df['message_date'],format='%d/%m/%Y, %H:%M - ')
df.rename(columns={'message_date':'date'},inplace=True)#inplace=true ni likhogi toh change ni krega
df.head()
```

```python
df.shape#to check how many mssgs have came
```

```python
# separate users and messages
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]:#user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user']=users
df['message']=messages
df.drop(columns=['user_message'],inplace=True)
df.head()


'''([\w\W]+?)
[...] → Capturing group (captures the username).
\w → Matches any word character (letters, digits, underscore).
\W → Matches any non-word character (spaces, punctuation, etc.).
\w\W → This combination ensures that ALL characters are matched.
+? → This is a lazy quantifier:
+ → Match one or more characters.
? → Lazy match (stops at the first : rather than matching everything).
:\s
: → Matches the colon (:) that separates the username from the message.
\s → Matches a whitespace character (space, tab, etc.), ensuring that the colon is followed by a space.'''
```

```python
# now using hour we made a new column period like if in hour its 16 so in period it will be 16-17
period=[]
for hour in df['hour']:
    if hour==23:
        period.append(str(hour)+"-"+str("00"))
    elif hour==0:
        period.append(str("00")+"-"+str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
print(period)
```

```python
df['period']=period
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))#figure size to 20 inches wide and 6 inches tall
sns.heatmap(df.pivot_table(index='day_name',columns='period',values='message',aggfunc='count').fillna(0))#jhan par koi values ni hogi vha na hota toh us na ko hmne 0 se replace kr diya
plt.yticks(rotation="horizontal")
plt.show()
print(df.pivot_table(index='day_name',columns='period',values='message',aggfunc='count').fillna(0))

'''index='day_name': Rows represent different days of the week.
columns='period': Columns represent different time periods.
values='message': We are counting the occurrences of messages.
aggfunc='count': It counts the number of messages for each combination of day_name and period.
.fillna(0): If there are missing values (i.e., no messages recorded for a specific day and period), they are replaced with 0 so the heatmap does not have empty spaces.'''
```
